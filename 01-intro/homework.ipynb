```python
df['duration'] = (df['tpep_dropoff_datetime']-df['tpep_pickup_datetime'])

```

```python
df["duration"] = pd.to_timedelta(df["duration"])

df["minutes"] = df["duration"].dt.components.minutes
```

```python

criteria = (df['minutes'] > 1) & (df['minutes']< 60)
data = df[ criteria] 
```

```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer

# Ensure the correct dataframe variable name is used
data['PULocationID'] = data['PULocationID'].astype(str)
data['DOLocationID'] = data['DOLocationID'].astype(str)

# Convert DataFrame to a list of dictionaries
data_dicts = data[['PULocationID', 'DOLocationID']].to_dict(orient='records')

# Apply Dictionary Vectorizer (keep sparse format)
dv = DictVectorizer(sparse=True)  # Default is True, but explicitly stating it
feature_matrix = dv.fit_transform(data_dicts)

# Get the dimensionality
num_columns = feature_matrix.shape[1]
print(f"Number of features (columns): {num_columns}")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Define target variable (response variable)
y = data['minutes'].values  # Ensure 'duration' column exists

# Train a Linear Regression model
lr = LinearRegression()
lr.fit(feature_matrix, y)

# Predict on the training data
y_pred = lr.predict(feature_matrix)

# Calculate RMSE
rmse = mean_squared_error(y, y_pred, squared=False)
print(f"RMSE on training data: {rmse:.2f}")
```

```python
validation_data = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet') 
```

```python
# Load validation dataset (Ensure it's preprocessed similarly)
validation_data['PULocationID'] = validation_data['PULocationID'].astype(str)
validation_data['DOLocationID'] = validation_data['DOLocationID'].astype(str)

# Convert validation set to dictionary format
val_data_dicts = validation_data[['PULocationID', 'DOLocationID']].to_dict(orient='records')

# Transform using the **same fitted DictVectorizer**
X_val = dv.transform(val_data_dicts)  

# Define target variable for validation set
y_val = validation_data['duration'].values  

# Predict on validation data
y_val_pred = lr.predict(X_val)

# Compute RMSE on validation
val_rmse = mean_squared_error(y_val, y_val_pred, squared=False)
print(f"RMSE on validation data: {val_rmse:.2f}")
```
